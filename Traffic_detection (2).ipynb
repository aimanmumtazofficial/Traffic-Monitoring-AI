```python
import os
os.makedirs('/content/utils', exist_ok=True)

# Speed file
with open('/content/utils/__init__.py', 'w') as f:
    f.write('')

with open('/content/utils/speed.py', 'w') as f:
    f.write('''
import numpy as np

class SpeedEstimator:
    def __init__(self):
        self.ppm = 8.0
        self.fps = 30
        self.tracks = {}

    def update(self, track_id, bbox):
        cx = int((bbox[0] + bbox[2]) / 2)
        cy = int((bbox[1] + bbox[3]) / 2)

        if track_id not in self.tracks:
            self.tracks[track_id] = {"positions": [(cx, cy)], "speeds": []}
        else:
            self.tracks[track_id]["positions"].append((cx, cy))
            if len(self.tracks[track_id]["positions"]) >= 2:
                p1 = self.tracks[track_id]["positions"][-2]
                p2 = self.tracks[track_id]["positions"][-1]
                dist_pixels = np.sqrt((p2[0]-p1[0])**2 + (p2[1]-p1[1])**2)
                dist_meters = dist_pixels / self.ppm
                speed = dist_meters * self.fps * 3.6
                self.tracks[track_id]["speeds"].append(speed)

        if self.tracks[track_id]["speeds"]:
            return min(np.mean(self.tracks[track_id]["speeds"][-5:]), 200)
        return 0.0

    def reset(self):
        self.tracks = {}
''')

with open('/content/utils/violation.py', 'w') as f:
    f.write('''
import pandas as pd
from datetime import datetime

class ViolationTracker:
    def __init__(self):
        self.violations = []
        self.signal_line_y = None
        self.signal_red = False
        self.recorded_ids = set()

    def set_signal_line(self, y_position):
        self.signal_line_y = y_position

    def set_signal_red(self, is_red):
        self.signal_red = is_red

    def check_violation(self, track_id, bbox, plate_text=""):
        if self.signal_line_y is None:
            return False
        cy = int((bbox[1] + bbox[3]) / 2)
        if self.signal_red and cy > self.signal_line_y and track_id not in self.recorded_ids:
            self.recorded_ids.add(track_id)
            self.violations.append({
                "Time": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                "Vehicle_ID": track_id,
                "Number_Plate": plate_text if plate_text else "Not Detected",
                "Violation": "Signal Jump"
            })
            return True
        return False

    def save_to_excel(self, filename="violations.xlsx"):
        if self.violations:
            df = pd.DataFrame(self.violations)
            df.to_excel(filename, index=False)
            return filename
        return None

    def get_violation_count(self):
        return len(self.violations)

    def reset(self):
        self.violations = []
        self.recorded_ids = set()
        self.signal_line_y = None
        self.signal_red = False
''')

print("Utils files created!")
```

```python
with open('/content/app.py', 'w') as f:
    f.write('''
import streamlit as st
import cv2
import tempfile
import numpy as np
from ultralytics import YOLO
import sys
sys.path.append("/content")
from utils.speed import SpeedEstimator
from utils.violation import ViolationTracker
import pandas as pd
import easyocr
import re

st.set_page_config(
    page_title="Traffic Monitoring System",
    page_icon="🚦",
    layout="wide"
)

# --- Header ---
st.title("🚦 AI Traffic Monitoring System")
st.markdown("Real-time vehicle detection, speed estimation, and violation monitoring")
st.markdown("---")

@st.cache_resource
def load_model():
    return YOLO("/content/best.pt")

@st.cache_resource
def load_ocr():
    return easyocr.Reader(["en"], gpu=False)

CLASS_NAMES = {
    0: "Unknown", 1: "Heavy_Bike", 2: "Bicycle", 3: "Bus",
    4: "Car", 5: "License_Plate", 6: "Motorbike", 7: "Helmet",
    8: "Truck", 9: "Rickshaw", 10: "Oil_Tanker", 11: "Truck"
}

VEHICLE_CLASSES = [1, 2, 3, 4, 6, 8, 9, 10, 11]

COLORS = {
    "Car": (0,255,0), "Bus": (255,165,0), "Truck": (255,0,0),
    "Rickshaw": (0,165,255), "Motorbike": (255,255,0),
    "Bicycle": (0,255,255), "Heavy_Bike": (255,0,255),
    "Oil_Tanker": (128,0,128), "Helmet": (0,128,255),
    "License_Plate": (255,128,0), "Unknown": (128,128,128)
}

# --- Sidebar ---
st.sidebar.image("https://img.icons8.com/color/96/traffic-light.png", width=80)
st.sidebar.title("⚙️ Control Panel")
st.sidebar.markdown("---")

st.sidebar.subheader("🎯 Detection Settings")
confidence = st.sidebar.slider("Confidence Threshold", 0.1, 1.0, 0.25, 0.05)

st.sidebar.markdown("---")
st.sidebar.subheader("🚦 Signal Settings")
signal_red = st.sidebar.toggle("🔴 Red Signal Active", value=False)
signal_line_pct = st.sidebar.slider("Signal Line Position (%)", 10, 90, 50)

# --- Load Models ---
model = load_model()
ocr_reader = load_ocr()
speed_estimator = SpeedEstimator()
violation_tracker = ViolationTracker()

# --- Live Statistics ---
st.subheader("📊 Live Statistics")
col1, col2, col3, col4, col5 = st.columns(5)
vehicle_count_ph = col1.empty()
violation_count_ph = col2.empty()
speed_ph = col3.empty()
helmet_ph = col4.empty()
plate_ph = col5.empty()

vehicle_count_ph.metric("🚗 Vehicles", "0")
violation_count_ph.metric("⚠️ Violations", "0")
speed_ph.metric("⚡ Max Speed", "0 km/h")
helmet_ph.metric("⛑️ No Helmet", "0")
plate_ph.metric("🔖 Plates Found", "0")

st.markdown("---")

# --- Video Upload ---
st.subheader("📹 Upload Traffic Video")
uploaded_file = st.file_uploader("Choose a video file", type=["mp4", "avi", "mov", "mkv"])

if uploaded_file:
    tfile = tempfile.NamedTemporaryFile(delete=False, suffix=".mp4")
    tfile.write(uploaded_file.read())
    tfile.flush()

    if st.button("🚀 Start Detection", type="primary", use_container_width=True):
        speed_estimator.reset()
        violation_tracker.reset()
        violation_tracker.set_signal_red(signal_red)

        cap = cv2.VideoCapture(tfile.name)
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        fps = cap.get(cv2.CAP_PROP_FPS)
        speed_estimator.fps = fps if fps > 0 else 30

        # UI Placeholders
        alert_ph = st.empty()
        frame_ph = st.empty()
        progress_bar = st.progress(0)
        status_text = st.empty()

        # Plate section placeholder (Initially empty)
        plates_header_ph = st.empty()
        plates_table_ph = st.empty()

        vehicle_ids = set()
        max_speed = 0
        no_helmet_count = 0
        detected_plates = []
        frame_num = 0

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret: break

            frame_num += 1
            h, w = frame.shape[:2]
            signal_y = int(h * signal_line_pct / 100)
            violation_tracker.set_signal_line(signal_y)

            # Process Frame
            results = model.track(frame, conf=confidence, persist=True, verbose=False, imgsz=640)

            if results[0].boxes is not None:
                for box in results[0].boxes:
                    if box.id is None: continue
                    track_id = int(box.id.item())
                    cls_id = int(box.cls.item())
                    bbox = box.xyxy[0].cpu().numpy()
                    class_name = CLASS_NAMES.get(cls_id, "Unknown")
                    x1, y1, x2, y2 = map(int, bbox)

                    # OCR Logic
                    plate_text = ""
                    if cls_id == 5:
                        try:
                            plate_crop = frame[y1:y2, x1:x2]
                            if plate_crop.size > 0:
                                gray_plate = cv2.cvtColor(plate_crop, cv2.COLOR_BGR2GRAY)
                                ocr_results = ocr_reader.readtext(gray_plate)
                                if ocr_results:
                                    raw_string = "".join([res[1] for res in ocr_results if res[2] > 0.3])
                                    clean_string = re.sub(r'[^A-Z0-9]', '', raw_string.upper())
                                    if len(clean_string) >= 3:
                                        plate_text = clean_string
                                        if plate_text not in [p["Plate"] for p in detected_plates]:
                                            detected_plates.append({"Vehicle_ID": track_id, "Plate": plate_text, "Frame": frame_num})
                        except: pass

                    # Violations & Drawing
                    is_v = violation_tracker.check_violation(track_id, bbox, plate_text)
                    box_color = (0,0,255) if is_v else COLORS.get(class_name, (128,128,128))
                    cv2.rectangle(frame, (x1, y1), (x2, y2), box_color, 2)
                    cv2.putText(frame, f"{class_name} #{track_id}", (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 2)

                    if cls_id in VEHICLE_CLASSES: vehicle_ids.add(track_id)

            # Update Live Table for Plates (Header appears only when data exists)
            if detected_plates:
                plates_header_ph.markdown("### 🔖 Detected Number Plates")
                plates_table_ph.dataframe(pd.DataFrame(detected_plates), use_container_width=True)

            # Update Frame & Metrics
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frame_ph.image(frame_rgb, use_container_width=True)
            vehicle_count_ph.metric("🚗 Vehicles", len(vehicle_ids))
            plate_ph.metric("🔖 Plates Found", len(detected_plates))

            progress = frame_num / total_frames if total_frames > 0 else 0
            progress_bar.progress(min(progress, 1.0))
            status_text.text(f"Processing frame {frame_num}/{total_frames}...")

        cap.release()
        st.success("✅ Detection process successfully completed!")
''')
print("Updated app.py with Dynamic Plates Header created!")
```

```python
# Web Scrapping:
scraping_code = '''
import requests
import os
import time
from urllib.parse import quote

def scrape_images_unsplash(query, num_images=50, save_folder="dataset"):
    os.makedirs(save_folder, exist_ok=True)
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
    downloaded = 0
    page = 1
    while downloaded < num_images:
        url = f"https://unsplash.com/napi/search/photos?query={quote(query)}&per_page=20&page={page}"
        try:
            response = requests.get(url, headers=headers, timeout=10)
            data = response.json()
            if "results" not in data:
                break
            for photo in data["results"]:
                if downloaded >= num_images:
                    break
                img_url = photo["urls"]["regular"]
                img_response = requests.get(img_url, timeout=10)
                filename = f"{save_folder}/{query.replace(' ', '_')}_{downloaded+1}.jpg"
                with open(filename, "wb") as f:
                    f.write(img_response.content)
                downloaded += 1
                print(f"Downloaded: {downloaded}/{num_images}")
                time.sleep(0.5)
            page += 1
        except Exception as e:
            print(f"Error: {e}")
            break
    print(f"Total downloaded: {downloaded} images for {query}")
    return downloaded

def scrape_all_classes():
    classes = {
        "car traffic road": 50,
        "motorcycle road": 50,
        "bicycle traffic": 50,
        "rickshaw auto india": 50,
        "truck highway": 50,
        "oil tanker truck": 50,
        "heavy bike motorcycle": 50,
    }
    total = 0
    for query, count in classes.items():
        print(f"Scraping: {query}")
        downloaded = scrape_images_unsplash(query, count, f"dataset/{query.split()[0]}")
        total += downloaded
    print(f"Total images collected: {total}")

if __name__ == "__main__":
    scrape_all_classes()
'''

with open('/content/traffic_project/scripts/scrape_images.py', 'w') as f:
    f.write(scraping_code)

print("Scraping script created!")
```

```python
#Inference Script:
inference_code = '''
from ultralytics import YOLO
import cv2

CLASS_NAMES = {
    0: "Unknown", 1: "Heavy_Bike", 2: "Bicycle", 3: "Bus",
    4: "Car", 5: "License_Plate", 6: "Motorbike", 7: "Helmet",
    8: "Truck", 9: "Rickshaw", 10: "Oil_Tanker", 11: "Truck"
}

COLORS = {
    "Car": (0,255,0), "Bus": (255,165,0), "Truck": (255,0,0),
    "Rickshaw": (0,165,255), "Motorbike": (255,255,0),
    "Bicycle": (0,255,255), "Heavy_Bike": (255,0,255),
    "Oil_Tanker": (128,0,128), "Helmet": (0,128,255),
    "License_Plate": (255,128,0), "Unknown": (128,128,128)
}

def run_inference(video_path, model_path="best.pt", output_path="output.mp4"):
    model = YOLO(model_path)
    cap = cv2.VideoCapture(video_path)
    w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    out = cv2.VideoWriter(output_path, fourcc, fps, (w, h))
    vehicle_ids = set()
    frame_num = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        frame_num += 1
        results = model.track(frame, conf=0.5, persist=True, verbose=False)
        if results[0].boxes is not None:
            for box in results[0].boxes:
                if box.id is None:
                    continue
                track_id = int(box.id.item())
                cls_id = int(box.cls.item())
                bbox = box.xyxy[0].cpu().numpy()
                class_name = CLASS_NAMES.get(cls_id, "Unknown")
                color = COLORS.get(class_name, (128,128,128))
                x1,y1,x2,y2 = int(bbox[0]),int(bbox[1]),int(bbox[2]),int(bbox[3])
                vehicle_ids.add(track_id)
                cv2.rectangle(frame, (x1,y1), (x2,y2), color, 2)
                cv2.putText(frame, f"{class_name} {track_id}",
                    (x1, y1-5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
        cv2.putText(frame, f"Vehicles: {len(vehicle_ids)}",
            (10,30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,255), 2)
        out.write(frame)
        print(f"Frame {frame_num} done", end="\\r")
    cap.release()
    out.release()
    print(f"\\nTotal vehicles: {len(vehicle_ids)}")
    print(f"Output: {output_path}")

if __name__ == "__main__":
    run_inference("traffic_video.mp4", "best.pt", "output_detected.mp4")
'''

with open('/content/traffic_project/inference.py', 'w') as f:
    f.write(inference_code)

print("Inference script created!")
```

```python
with open('/content/app.py', 'w') as f:
    f.write('''
import streamlit as st
import cv2
import tempfile
import numpy as np
from ultralytics import YOLO
import sys
sys.path.append("/content")
from utils.speed import SpeedEstimator
from utils.violation import ViolationTracker
import pandas as pd
import easyocr
import re

st.set_page_config(
    page_title="Traffic Monitoring System",
    page_icon="🚦",
    layout="wide"
)

# --- Header ---
st.title("🚦 AI Traffic Monitoring System")
st.markdown("Real-time vehicle detection, speed estimation, and violation monitoring")
st.markdown("---")

@st.cache_resource
def load_model():
    return YOLO("/content/best.pt")

@st.cache_resource
def load_ocr():
    return easyocr.Reader(["en"], gpu=False)

CLASS_NAMES = {
    0: "Unknown", 1: "Heavy_Bike", 2: "Bicycle", 3: "Bus",
    4: "Car", 5: "License_Plate", 6: "Motorbike", 7: "Helmet",
    8: "Truck", 9: "Rickshaw", 10: "Oil_Tanker", 11: "Truck"
}

VEHICLE_CLASSES = [1, 2, 3, 4, 6, 8, 9, 10, 11]

COLORS = {
    "Car": (0,255,0), "Bus": (255,165,0), "Truck": (255,0,0),
    "Rickshaw": (0,165,255), "Motorbike": (255,255,0),
    "Bicycle": (0,255,255), "Heavy_Bike": (255,0,255),
    "Oil_Tanker": (128,0,128), "Helmet": (0,128,255),
    "License_Plate": (255,128,0), "Unknown": (128,128,128)
}

# --- Sidebar ---
st.sidebar.image("https://img.icons8.com/color/96/traffic-light.png", width=80)
st.sidebar.title("⚙️ Control Panel")
st.sidebar.markdown("---")

st.sidebar.subheader("🎯 Detection Settings")
confidence = st.sidebar.slider("Confidence Threshold", 0.1, 1.0, 0.25, 0.05)

st.sidebar.markdown("---")
st.sidebar.subheader("🚦 Signal Settings")
signal_red = st.sidebar.toggle("🔴 Red Signal Active", value=False)
signal_line_pct = st.sidebar.slider("Signal Line Position (%)", 10, 90, 50)

# --- Load Models ---
model = load_model()
ocr_reader = load_ocr()
speed_estimator = SpeedEstimator()
violation_tracker = ViolationTracker()

# --- Live Statistics ---
st.subheader("📊 Live Statistics")
col1, col2, col3, col4, col5 = st.columns(5)
vehicle_count_ph = col1.empty()
violation_count_ph = col2.empty()
speed_ph = col3.empty()
helmet_ph = col4.empty()
plate_ph = col5.empty()

vehicle_count_ph.metric("🚗 Vehicles", "0")
violation_count_ph.metric("⚠️ Violations", "0")
speed_ph.metric("⚡ Max Speed", "0 km/h")
helmet_ph.metric("⛑️ No Helmet", "0")
plate_ph.metric("🔖 Plates Found", "0")

st.markdown("---")

# --- Video Upload ---
st.subheader("📹 Upload Traffic Video")
uploaded_file = st.file_uploader("Choose a video file", type=["mp4", "avi", "mov", "mkv"])

if uploaded_file:
    tfile = tempfile.NamedTemporaryFile(delete=False, suffix=".mp4")
    tfile.write(uploaded_file.read())
    tfile.flush()

    if st.button("🚀 Start Detection", type="primary", use_container_width=True):
        speed_estimator.reset()
        violation_tracker.reset()
        violation_tracker.set_signal_red(signal_red)

        cap = cv2.VideoCapture(tfile.name)
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        fps = cap.get(cv2.CAP_PROP_FPS)
        speed_estimator.fps = fps if fps > 0 else 30

        # UI Placeholders
        alert_ph = st.empty()
        frame_ph = st.empty()
        progress_bar = st.progress(0)
        status_text = st.empty()

        # Plate section placeholder (Initially empty)
        plates_header_ph = st.empty()
        plates_table_ph = st.empty()

        vehicle_ids = set()
        max_speed = 0
        no_helmet_count = 0
        detected_plates = []
        frame_num = 0

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret: break

            frame_num += 1
            h, w = frame.shape[:2]
            signal_y = int(h * signal_line_pct / 100)
            violation_tracker.set_signal_line(signal_y)

            # Process Frame
            results = model.track(frame, conf=confidence, persist=True, verbose=False, imgsz=640)

            if results[0].boxes is not None:
                for box in results[0].boxes:
                    if box.id is None: continue
                    track_id = int(box.id.item())
                    cls_id = int(box.cls.item())
                    bbox = box.xyxy[0].cpu().numpy()
                    class_name = CLASS_NAMES.get(cls_id, "Unknown")
                    x1, y1, x2, y2 = map(int, bbox)

                    # OCR Logic
                    plate_text = ""
                    if cls_id == 5:
                        try:
                            plate_crop = frame[y1:y2, x1:x2]
                            if plate_crop.size > 0:
                                gray_plate = cv2.cvtColor(plate_crop, cv2.COLOR_BGR2GRAY)
                                ocr_results = ocr_reader.readtext(gray_plate)
                                if ocr_results:
                                    raw_string = "".join([res[1] for res in ocr_results if res[2] > 0.3])
                                    clean_string = re.sub(r'[^A-Z0-9]', '', raw_string.upper())
                                    if len(clean_string) >= 3:
                                        plate_text = clean_string
                                        if plate_text not in [p["Plate"] for p in detected_plates]:
                                            detected_plates.append({"Vehicle_ID": track_id, "Plate": plate_text, "Frame": frame_num})
                        except: pass

                    # Violations & Drawing
                    is_v = violation_tracker.check_violation(track_id, bbox, plate_text)
                    box_color = (0,0,255) if is_v else COLORS.get(class_name, (128,128,128))
                    cv2.rectangle(frame, (x1, y1), (x2, y2), box_color, 2)
                    cv2.putText(frame, f"{class_name} #{track_id}", (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 2)

                    if cls_id in VEHICLE_CLASSES: vehicle_ids.add(track_id)

            # Update Live Table for Plates (Header appears only when data exists)
            if detected_plates:
                plates_header_ph.markdown("### 🔖 Detected Number Plates")
                plates_table_ph.dataframe(pd.DataFrame(detected_plates), use_container_width=True)

            # Update Frame & Metrics
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frame_ph.image(frame_rgb, use_container_width=True)
            vehicle_count_ph.metric("🚗 Vehicles", len(vehicle_ids))
            plate_ph.metric("🔖 Plates Found", len(detected_plates))

            progress = frame_num / total_frames if total_frames > 0 else 0
            progress_bar.progress(min(progress, 1.0))
            status_text.text(f"Processing frame {frame_num}/{total_frames}...")

        cap.release()
        st.success("✅ Detection process successfully completed!")
''')
print("Updated app.py with Dynamic Plates Header created!")
```
